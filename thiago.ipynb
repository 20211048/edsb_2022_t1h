Torna-se necessÃ¡rio transformar a feature 'volume_primary_units' numÃ©rica em 'float' pois estÃ£o como 'object'.

```python
df['volume_primary_units'] = df['volume_primary_units'].replace(',', '.', regex=True).astype(float)
#df['volume_primary_units'] = df['volume_primary_units'].replace('\,\d+', '', regex=True).astype(float)
#jÃ¡ nÃ£o Ã© necessÃ¡rio pois importou-se a "," como decimal 
#df['invoiced_sales'] = df['invoiced_sales'].replace('\,\d+', '', regex=True).astype(float)
#df['inventory_cost'] = df['inventory_cost'].replace('\,\d+', '', regex=True).astype(float)
df
```

Alterar formato do 'month_code' para ano e mÃªs.

```python
df['month_code'] = pd.to_datetime(df['month_code'].astype(float).astype(int), format='%Y%m')
```

Conclui-se que dos cerca de 286 mil a grande maioria das datas em 1900 tambÃ©m tem valores numÃ©ricos em 0, e portanto removeremos todos pois nÃ£o representa grande perda no dataset com mais de 4.8 milhÃµes de entradas.

```python
df = df.loc[df['month_code']!='1900-01-01']
df.reset_index(drop=True,inplace=True)
df
```
