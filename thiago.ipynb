```python
df = df.dropna()
df
```

Torna-se necessário transformar a feature 'volume_primary_units' numérica em 'float' pois estão como 'object'.

```python
df['volume_primary_units'] = df['volume_primary_units'].replace(',', '.', regex=True).astype(float)
#df['volume_primary_units'] = df['volume_primary_units'].replace('\,\d+', '', regex=True).astype(float)
#já não é necessário pois importou-se a "," como decimal 
#df['invoiced_sales'] = df['invoiced_sales'].replace('\,\d+', '', regex=True).astype(float)
#df['inventory_cost'] = df['inventory_cost'].replace('\,\d+', '', regex=True).astype(float)
df
```

Alterar formato do 'month_code' para ano e mês.

```python
df['month_code'] = pd.to_datetime(df['month_code'].astype(float).astype(int), format='%Y%m')
```

O valor de data em '1900-01-01' é estranho e indica que não foi preenchido, portanto vamos verificar o que tem preenchido nos outros campos:

```python
df[df['month_code']=='1900-01-01'].describe()
```

```python
df[df['month_code']=='1900-01-01']['invoiced_sales'].value_counts()
```

```python
df[df['month_code']=='1900-01-01']['volume_primary_units'].value_counts()
```

```python
df[df['month_code']=='1900-01-01']['inventory_cost'].value_counts()
```

Conclui-se que dos cerca de 286 mil a grande maioria das datas em 1900 também tem valores numéricos em 0, e portanto removeremos todos pois não representa grande perda no dataset com mais de 4.8 milhões de entradas.

```python
df = df.loc[df['month_code']!='1900-01-01']
df.reset_index(drop=True,inplace=True)
df
```
